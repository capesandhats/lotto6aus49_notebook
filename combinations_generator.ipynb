## inputs

```python
#high and low division limit (limit number is considered as low number)
low_number_till=30

#last_3_results=[2,22,25,32,45,46,7,11,31,34,35,36,8,9,39,40,41,43]
last_5_results=[2,22,25,32,45,46,7,11,31,34,35,36,8,9,39,40,41,43,1,6,7,20,29,49,5,14,19,22,33,41]
results=last_5_results

# filter even and high values combinations choose 

even_values_combi=[3,2,4]
high_values_combi=[4,3,5]

# filter the diag combinations
diag_combi=[0,1,2]

print("the input results are stored in the variable : results")
```

```python
print("input:"+  str(results))
results=np.sort(results)
results = np.unique(results)
print("sorted: " + str(results))
print("length: " + str(len(results)))

even = [i for i in results if i%2==0]
odd  = [i for i in results if i%2!=0]
high = [i for i in results if i>low_number_till]
low  = [i for i in results if i<=low_number_till]
print("even: " + str(even))
print("odd: " + str(odd))
print("high: " + str(high))
print("low: " + str(low))
```

## combinations

```python
import itertools
start_time = timeit.default_timer()
total_combinations = list(itertools.combinations(results, 6))
print("total number of possible combinations: " + str(len(total_combinations)))

df_total_combinations=pd.DataFrame(total_combinations)
df_total_combinations=df_total_combinations.rename(index=str, columns={0: "n1", 1: "n2", 2: "n3", 3: "n4", 4: "n5", 5: "n6"})

for index, row in df_total_combinations.iterrows() :
    #print(str(index) + str(row))
    even = [i for i in row[0:6] if i%2==0]
    odd  = [i for i in row[0:6] if i%2!=0]
    high = [i for i in row[0:6] if i>low_number_till]
    low  = [i for i in row[0:6] if i<=low_number_till]
    #print(row[0:6])
    #print(high)
    #print(len(high))
    #print(low)
    #display(row)
    df_total_combinations.at[index,'even'] = np.count_nonzero(even)
    df_total_combinations.at[index,'odd']  = np.count_nonzero(odd)
    df_total_combinations.at[index,'high'] = len(high)
    df_total_combinations.at[index,'low']  = len(low)
print("shape of combinations:" +str(df_total_combinations.shape))
elapsed = timeit.default_timer() - start_time
print("time taken to generate the combinations and compute even and odd: " + str(elapsed))
display(df_total_combinations.head(3))
display(df_total_combinations.tail(3))
```
