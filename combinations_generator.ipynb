## load past results

```python
DIR_NAME = os.getcwd()

start_time = timeit.default_timer()

df2018=pd.read_csv(os.path.join(DIR_NAME, 'lottodata/2018.txt'))
df2017=pd.read_csv(os.path.join(DIR_NAME, 'lottodata/2017.txt'))
df2016=pd.read_csv(os.path.join(DIR_NAME, 'lottodata/2016.txt'))

#fromating data

dt2018 =df2018.rename(columns={'Date': 'date', 'number1' : 'n1', 'number2':'n2', 'number3':'n3', 'number4':'n4', 'number5':'n5',
       'number6':'n6', 'number7':'n7', 'jackpot clamied': 'jackpot'}, index ={})
dt2018.date= pd.to_datetime(dt2018.date)

dt2017 =df2017.rename(columns={'Date': 'date', 'number1' : 'n1', 'number2':'n2', 'number3':'n3', 'number4':'n4', 'number5':'n5',
       'number6':'n6', 'number7':'n7', 'jackpot clamied': 'jackpot'}, index ={})
dt2017.date= pd.to_datetime(dt2017.date)

dt2016 =df2016.rename(columns={'Date': 'date', 'number1' : 'n1', 'number2':'n2', 'number3':'n3', 'number4':'n4', 'number5':'n5',
       'number6':'n6', 'number7':'n7', 'jackpot clamied': 'jackpot'}, index ={})
dt2016.date= pd.to_datetime(dt2016.date)

frames = [dt2018,dt2017,dt2016]
dt = pd.concat(frames)
dt.reset_index(inplace=True, drop=True) 
print("Loaded data from :", dt.iloc[dt.date.size-1][0]," to :", dt.iloc[0][0])

elapsed = timeit.default_timer() - start_time

print("time elapsed to load data: "+ str(elapsed))
display(dt.head(5))
```

## inputs

```python
# computing the past results
no_past_results =6
#---------------------


#high and low division limit (limit number is considered as low number)
low_number_till=30
#------------------

# filter the maximum numbers of past results can occur 
past_combi =[0,1,2]
#-------------------

# filter even and high values combinations choose 

even_values_combi=[3,2,4]
high_values_combi=[4,3,5]
#------------------------------

# filter the diag combinations
diag_combi=[0,1,2]
#------------------------------

print("the input results are stored in the variable : results")

last_i_results= dt.iloc[ 1 : no_past_results+1, 1:7].values
last_results = np.unique(last_i_results.flatten())
print(last_results)
print("last results unique combinations: \n" + str(last_results))
print("last results unique combinations size: "+ str(last_results.size))
#last_3_results=[2,22,25,32,45,46,7,11,31,34,35,36,8,9,39,40,41,43]
#last_5_results=[2,22,25,32,45,46,7,11,31,34,35,36,8,9,39,40,41,43,1,6,7,20,29,49,5,14,19,22,33,41]
results=last_i_results
```

```python
print("input:"+  str(results))
results=np.sort(results)
results = np.unique(results)
print("sorted: " + str(results))
print("length: " + str(len(results)))

even = [i for i in results if i%2==0]
odd  = [i for i in results if i%2!=0]
high = [i for i in results if i>low_number_till]
low  = [i for i in results if i<=low_number_till]
print("even: " + str(even))
print("odd: " + str(odd))
print("high: " + str(high))
print("low: " + str(low))
```
