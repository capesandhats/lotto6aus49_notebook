## Read past results

```python
INITIAL_COUNT = 57
COUNTERS_DF = pd.DataFrame()
FILTERED_DF = pd.DataFrame()
FREQUENCY_ORDER_OF_NUMBERS =""
DIR_NAME = os.getcwd()

start_time = timeit.default_timer()

df2018=pd.read_csv(os.path.join(DIR_NAME, 'lottodata/2018.txt'))
df2017=pd.read_csv(os.path.join(DIR_NAME, 'lottodata/2017.txt'))
df2016=pd.read_csv(os.path.join(DIR_NAME, 'lottodata/2016.txt'))

#fromating data

dt2018 =df2018.rename(columns={'Date': 'date', 'number1' : 'n1', 'number2':'n2', 'number3':'n3', 'number4':'n4', 'number5':'n5',
       'number6':'n6', 'number7':'n7', 'jackpot clamied': 'jackpot'}, index ={})
dt2018.date= pd.to_datetime(dt2018.date)

dt2017 =df2017.rename(columns={'Date': 'date', 'number1' : 'n1', 'number2':'n2', 'number3':'n3', 'number4':'n4', 'number5':'n5',
       'number6':'n6', 'number7':'n7', 'jackpot clamied': 'jackpot'}, index ={})
dt2017.date= pd.to_datetime(dt2017.date)

dt2016 =df2016.rename(columns={'Date': 'date', 'number1' : 'n1', 'number2':'n2', 'number3':'n3', 'number4':'n4', 'number5':'n5',
       'number6':'n6', 'number7':'n7', 'jackpot clamied': 'jackpot'}, index ={})
dt2016.date= pd.to_datetime(dt2016.date)

frames = [dt2018,dt2017,dt2016]
dt = pd.concat(frames)
dt.reset_index(inplace=True, drop=True) 
print("Loaded data from :", dt.iloc[dt.date.size-1][0]," to :", dt.iloc[0][0])

elapsed = timeit.default_timer() - start_time

print("time elapsed to load data: "+ str(elapsed))
display(dt.head(5))
```

## computing even and odd 

```python
low_number_till=30
print("High number is from: " + str(low_number_till))
for index, row in dt.iterrows() :
    #print(row[1:7])
    #print(str(index) + str(row))
    even = [i for i in row[1:7] if i%2==0]
    odd  = [i for i in row[1:7] if i%2!=0]
    high = [i for i in row[1:7] if i>low_number_till]
    low  = [i for i in row[1:7] if i<=low_number_till]
    #print(row[0:6])
    #print(high)
    #print(len(high))
    #print(low)
    #display(row)
    dt.at[index,'even'] = np.count_nonzero(even)
    dt.at[index,'odd']  = np.count_nonzero(odd)
    dt.at[index,'high'] = len(high)
    dt.at[index,'low']  = len(low)
print("shape of combinations ""dt"": " +str(dt.shape))
elapsed = timeit.default_timer() - start_time
print("time elapsed to do even and odd analaysis: "+ str(elapsed))
#dt.iloc[:,1:7].head(10)
display(dt.head(5))
```
